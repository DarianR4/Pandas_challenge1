```python
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "./Resources/schools_complete.csv"
student_data_to_load = "./Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#confirm columns
school_data_complete
```

```python
#District Summary

#calculation of total students, total budget, total schools, Avg Math score, Avg reading score, 
total_students = len(school_data_complete["student_name"])
total_budget = school_data_complete["budget"].sum()
total_schools = len(school_data_complete["school_name"].unique())
avg_mathscore = school_data_complete["math_score"].mean()
avg_reading = school_data_complete["reading_score"].mean()
math_passers = len(school_data_complete[school_data_complete.math_score > 70])
read_passers = len(school_data_complete[school_data_complete.reading_score > 70])
math_pass_pct = (math_passers/total_students) *100
read_pass_pct = (read_passers/total_students) *100
     
passing_math_reading = len(school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)])
passing_math_reading_pct = (passing_math_reading/total_students)*100 

# schoolgroup_data["% Passing Math"]= math_pass_pct
# schoolgroup_data["% Passing Reading"]= read_pass_pct
#schoolgroup_data["% Overall Passing"]= passing_math_reading_pct

passing_math_reading_pct


district_summary = pd.DataFrame({"Total Schools":[total_schools],'Total Students':[total_students],"Total Budget":[total_budget],
                    "Average Math Score":[avg_mathscore], "Average Reading Score":[avg_reading],
                    '% Passing Math':[math_pass_pct],'% Passing Reading':[read_pass_pct],'% Overal Passing':[passing_math_reading_pct]})
district_summary
```

```python
#school summary section

schoolgroup_types = school_data_complete.set_index(['school_name'])["type"]

schoolgroup_data = pd.DataFrame(schoolgroup_types)
schoolgroup_data = pd.DataFrame({"School Type": schoolgroup_types})
schoolgroup_data
```

```python
#total students
schoolgroup_counts = school_data_complete['school_name'].value_counts()
schoolgroup_counts
```

```python
#total school budget
schoolgroup_budget_gr = school_data_complete.groupby("school_name")
school_group_budget = schoolgroup_budget_gr["budget"].max()
school_group_budget
```

```python
buget_perstudent = school_group_budget/schoolgroup_counts

buget_perstudent
```

```python
#per student spending
#schoolgroup_psbudget = schoolgroup_budget/schoolgroup_counts

#passing scores by school

schoolgroup_mathscore = school_data_complete.loc[(school_data_complete["math_score"]>70)]

schoolgroup_readingscore = school_data_complete.loc[(school_data_complete["reading_score"]>70)]

#passing students and percentages
schoolgroup_mathscore = schoolgroup_mathscore.groupby(["school_name"]).count()["student_name"]
schoolgroup_readingscore = schoolgroup_readingscore.groupby(["school_name"]).count()["student_name"]

#average scores by school
schoolgroup_mathavg = schoolgroup_mathscore/schoolgroup_counts*100
schoolgroup_readingavg = schoolgroup_readingscore/schoolgroup_counts*100

# #overall passing

schoolgroup_overallp = school_data_complete.loc[(school_data_complete["math_score"]>70)&
                                            (school_data_complete["reading_score"]>70)]


schoolgroup_overallpz = schoolgroup_overallp.groupby(["school_name"]).count()

# # schoolgroup_overallpz
schoolgroup_bothpasspercent = schoolgroup_mathavg



schoolgroup_summary = pd.DataFrame({"Total Students":schoolgroup_counts,
                       "Budget":school_group_budget, "Average Math Score":schoolgroup_mathavg,"Average Reading Score": schoolgroup_readingavg,
                                    "Overall Passing %" : schoolgroup_bothpasspercent,"Budget Per Student":buget_perstudent})
schoolgroup_summary
```

```python
#Top performing school section

schoolgroup_summarysort = schoolgroup_summary.sort_values("Overall Passing %")
schoolgroup_summarysort.tail(5)
```

```python
#bottom performing schools

schoolgroup_summarysort = schoolgroup_summary.sort_values("Overall Passing %")
schoolgroup_summarysort.head(5)
```
