```python
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "./Resources/schools_complete.csv"
student_data_to_load = "./Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#confirm columns
school_data_complete
```

```python
#District Summary

#calculation of total students, total budget, total schools, Avg Math score, Avg reading score, 
total_students = len(school_data_complete["student_name"])
total_budget = school_data_complete["budget"].sum()
total_schools = len(school_data_complete["school_name"].unique())
avg_mathscore = school_data_complete["math_score"].mean()
avg_reading = school_data_complete["reading_score"].mean()
math_passers = len(school_data_complete[school_data_complete.math_score > 70])
read_passers = len(school_data_complete[school_data_complete.reading_score > 70])
math_pass_pct = (math_passers/total_students) *100
read_pass_pct = (read_passers/total_students) *100
     
passing_math_reading = len(school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)])
passing_math_reading_pct = (passing_math_reading/total_students)*100 

# schoolgroup_data["% Passing Math"]= math_pass_pct
# schoolgroup_data["% Passing Reading"]= read_pass_pct
#schoolgroup_data["% Overall Passing"]= passing_math_reading_pct

passing_math_reading_pct


district_summary = {"Total Schools":[total_schools],'Total Students':[total_students],"Total Budget":[total_budget],
                    "Average Math Score":[avg_mathscore], "Average Reading Score":[avg_reading],
                    '% Passing Math':[math_pass_pct],'% Passing Reading':[read_pass_pct],'% Overal Passing':[passing_math_reading_pct]}
district_summary
```

```python
#school summary section

schoolgroup_types = school_data_complete.set_index(['school_name'])["type"]

schoolgroup_data = pd.DataFrame(schoolgroup_types)
schoolgroup_data = pd.DataFrame({"School Type": schoolgroup_types})


#total students
schoolgroup_counts = school_data_complete['school_name'].value_counts()

#total school budget
schoolgroup_budget = school_data_complete.set_index(['school_name'])["budget"]


#per student spending
schoolgroup_psbudget = schoolgroup_budget/schoolgroup_counts

#passing scores by school

schoolgroup_mathscore = school_data_complete[(school_data_complete["math_score"]>70)]

schoolgroup_readingscore = school_data_complete[(school_data_complete["reading_score"]>70)]

#passing students and percentages
schoolgroup_mathscore = schoolgroup_mathscore.groupby(["school_name"]).count()["student_name"]
schoolgroup_readingscore = schoolgroup_readingscore.groupby(["school_name"]).count()["student_name"]

#average scores by school
schoolgroup_mathavg = schoolgroup_mathscore/schoolgroup_counts*100
schoolgroup_readingavg = schoolgroup_readingscore/schoolgroup_counts*100

# #overall passing

schoolgroup_overallp = len(school_data_complete[(school_data_complete["math_score"]>70)&
                                            (school_data_complete["reading_score"]>70)])

# schoolgroup_overallp

schoolgroup_bothpasspercent = (schoolgroup_overallp/schoolgroup_counts)*100

schoolgroup_bothpasspercent


# schoolgroup_summary = {"Schools":[schoolgroup_data],'School Type':[schoolgroup_types], 'Total Student':[schoolgroup_counts],'Total School Budget':[schoolgroup_budget],'Budget Per Student':[schoolgroup_psbudget],
#                       'Average Math Score':[schoolgroup_mathscore],'Average Reading Score':[schoolgroup_readingscore],"% Passing Math":[schoolgroup_mathavg],
#                       '% Passing Reading':[schoolgroup_readingavg],'% Overall Passing':[schoolgroup_bothpasspercent]}


```
